# RED Light Voilations


```python
import cv2
import numpy as np
from ultralytics import YOLO

# Load YOLO vehicle detection model
model = YOLO("yolov8n.pt")

# Video source (camera or file)
cap = cv2.VideoCapture(r"C:\Users\Parva mehta\Downloads\Road safety video _ Example of blind spot in truck car road accident...mp4")

# Define STOP LINE (y coordinate)
STOP_LINE_Y = 400

# Traffic light state (simulate)
traffic_light = "RED"   # change dynamically later

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()
        classes = result.boxes.cls.cpu().numpy()

        for box, cls in zip(boxes, classes):
            x1, y1, x2, y2 = map(int, box)

            # Vehicle classes in COCO
            if int(cls) in [2, 3, 5, 7]:  # car, bike, bus, truck

                cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)

                vehicle_bottom = y2

                # Check violation
                if traffic_light == "RED" and vehicle_bottom > STOP_LINE_Y:
                    cv2.putText(frame, "RED LIGHT VIOLATION",
                                (x1, y1-10),
                                cv2.FONT_HERSHEY_SIMPLEX,
                                0.7,
                                (0,0,255),
                                2)

    # Draw stop line
    cv2.line(frame, (0, STOP_LINE_Y), (frame.shape[1], STOP_LINE_Y), (0,0,255), 3)

    cv2.imshow("Red Light Detection", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
print(cap)
```

# Traffic Congestion

```python
import cv2
from ultralytics import YOLO
from matplotlib import pyplot as plt
from IPython.display import display, clear_output

model = YOLO("yolov8n.pt")

cap = cv2.VideoCapture(r"C:\Users\Parva mehta\Downloads\4K camera example for Traffic Monitoring Road.mp4")

frame_skip = 5   # process every 5th frame

frame_id = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_id += 1
    if frame_id % frame_skip != 0:
        continue

    # Reduce size heavily (important)
    frame = cv2.resize(frame, (480, 270))

    results = model(frame, verbose=False)

    annotated = results[0].plot()

    frame_rgb = cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB)

    clear_output(wait=True)
    display(plt.imshow(frame_rgb))
    plt.axis("off")
    plt.close()

cap.release()
```

# Wrong Way Driving

```python
import cv2
import numpy as np
from ultralytics import YOLO

# Load YOLO model
model = YOLO("yolov8n.pt")

# Input video
cap = cv2.VideoCapture(r"C:\Users\Parva mehta\Downloads\Man drives wrong way for 7.5 miles - CBS News (720p, h264) (1).mp4")

# Video properties
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = int(cap.get(cv2.CAP_PROP_FPS))

# Output video writer
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter("wrong_way_output.avi", fourcc, fps, (width, height))

# Allowed direction: vehicles should move LEFT  RIGHT
ALLOWED_DIRECTION = "RIGHT"

# Store previous centroids
previous_positions = {}

vehicle_classes = [2, 3, 5, 7]  # car, bike, bus, truck

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    current_positions = {}

    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()
        classes = result.boxes.cls.cpu().numpy()

        for i, (box, cls) in enumerate(zip(boxes, classes)):

            if int(cls) in vehicle_classes:
                x1, y1, x2, y2 = map(int, box)

                cx = int((x1 + x2) / 2)
                cy = int((y1 + y2) / 2)

                current_positions[i] = (cx, cy)

                # Draw bounding box
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)

                # Check direction
                if i in previous_positions:
                    prev_cx, prev_cy = previous_positions[i]

                    direction = "RIGHT" if cx > prev_cx else "LEFT"

                    if direction != ALLOWED_DIRECTION:
                        cv2.putText(frame, "WRONG WAY ALERT",
                                    (x1, y1 - 10),
                                    cv2.FONT_HERSHEY_SIMPLEX,
                                    0.7,
                                    (0,0,255),
                                    2)

    previous_positions = current_positions

    # Save output frame
    out.write(frame)

    # Display (works if running as .py)
    cv2.imshow("Wrong Way Detection", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2
import easyocr
from ultralytics import YOLO

# Vehicle detection model
vehicle_model = YOLO("yolov8n.pt")

# License plate detection model
plate_model = YOLO("license_plate_detector.pt")

# OCR reader
reader = easyocr.Reader(['en'])

# Input video
cap = cv2.VideoCapture(r"C:\Users\Parva mehta\Downloads\Toll plaza automatic vehicle recognition.mp4")

width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = int(cap.get(cv2.CAP_PROP_FPS))

# Output writer
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter("anpr_output.avi", fourcc, fps, (width, height))

vehicle_classes = [2,3,5,7]  # car, bike, bus, truck

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Vehicle detection
    vehicle_results = vehicle_model(frame)

    # Plate detection
    plate_results = plate_model(frame)

    # Draw vehicles
    for result in vehicle_results:
        boxes = result.boxes.xyxy.cpu().numpy()
        classes = result.boxes.cls.cpu().numpy()

        for box, cls in zip(boxes, classes):
            if int(cls) in vehicle_classes:
                x1, y1, x2, y2 = map(int, box)
                cv2.rectangle(frame, (x1,y1), (x2,y2), (0,255,0), 2)

    # Detect plates
    for result in plate_results:
        boxes = result.boxes.xyxy.cpu().numpy()

        for box in boxes:
            x1, y1, x2, y2 = map(int, box)

            plate_crop = frame[y1:y2, x1:x2]

            # OCR
            text = reader.readtext(plate_crop)
            plate_text = ""

            if len(text) > 0:
                plate_text = text[0][-2]

            cv2.rectangle(frame, (x1,y1), (x2,y2), (255,0,0), 2)
            cv2.putText(frame, plate_text,
                        (x1, y1-10),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.8,
                        (0,255,255),
                        2)

    # Save frame
    out.write(frame)

    cv2.imshow("ANPR Detection", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
